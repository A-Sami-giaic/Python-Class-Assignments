```python
# 1. Create and Modify Lists
fruits = ["apple", "banana", "orange"]
fruits.append("grape")              # Add "grape" to the end
fruits.insert(2, "mango")           # Insert "mango" at 2 index
fruits.remove("banana")             # Remove "banana"
print("Final fruits list:", fruits)

# 2. List Operations
numbers = [10, 20, 30, 40, 50]
more_numbers = [60, 70, 80]
all_numbers = numbers + more_numbers           # Combine lists
numbers_copy = all_numbers.copy()              # Copy list
numbers_copy.reverse()                         # Reverse copy
print("All numbers:", all_numbers)
print("Reversed copy:", numbers_copy)

# 3. Sorting and Counting
scores = [85, 92, 78, 65, 92, 85, 74]
scores_sorted_asc = sorted(scores)             # Sort ascending
scores_sorted_desc = sorted(scores, reverse=True)  # Sort descending
count_92 = scores.count(92)                    # Count 92
highest_score = max(scores)
lowest_score = min(scores)

print("Scores sorted (asc):", scores_sorted_asc)
print("Scores sorted (desc):", scores_sorted_desc)
print("Number of times 92 appears:", count_92)
print("Highest score:", highest_score)
print("Lowest score:", lowest_score)

# 4. List Manipulation
letters = ["a", "b", "c", "d", "e"]
first_three = letters[:3]                      # Slicing first 3
last_two = letters[-2:]                        # Slicing last 2
index_of_c = letters.index("c")                # Index of "c"
letters[3] = "z"                                # Replace "d" with "z"
print("First three letters:", first_three)
print("Last two letters:", last_two)
print("Index of 'c':", index_of_c)
print("Modified letters list:", letters)
```

```python
# 1.First i am making variables
purchase_amount = float(input("Enter the total purchase amount: $"))
is_loyalty_member = input("Are you a loyalty member? (True/False): ").strip().lower() == "true"
day_of_week = input("Enter the current day of the week (e.g., 'Monday'): ").strip().capitalize()
is_holiday_season = input("Is it currently holiday season? (True/False): ").strip().lower() == "true"

# Initialize discounts
discount = 0
discount_details = []


# Discount based on purchase amount
if purchase_amount > 100:
    discount = 0.10 * purchase_amount
    discount_details.append(f"10% discount on purchase over $100: -${discount:.2f}")
elif purchase_amount > 50:
    discount = 0.05 * purchase_amount
    discount_details.append(f"5% discount on purchase over $50: -${discount:.2f}")

# Loyalty member discount
if is_loyalty_member:
    loyalty_discount = 0.05 * purchase_amount
    discount += loyalty_discount
    discount_details.append(f"5% loyalty discount: -${loyalty_discount:.2f}")

# Weekend discount
if day_of_week in ["Saturday", "Sunday"]:
    weekend_discount = 0.02 * purchase_amount
    discount += weekend_discount
    discount_details.append(f"Weekend discount (Saturday/Sunday): -${weekend_discount:.2f}")

# Holiday season discount
if is_holiday_season:
    holiday_discount = 0.08 * purchase_amount
    discount += holiday_discount
    discount_details.append(f"Holiday season discount: -${holiday_discount:.2f}")

# 3. Calculate the final price
final_price = purchase_amount - discount

# Display results
print(f"\nOriginal purchase amount: ${purchase_amount:.2f}")
for detail in discount_details:
    print(detail)
print(f"Final price after all discounts: ${final_price:.2f}")
```

```python
#For loop
# Pattern 1: Counting Numbers
print("Pattern 1: Counting Numbers")
for i in range(1, 11):
    print(i, end=" ")
print()  # New line after the pattern

# Pattern 2: Multiplication Table for the number 5
print("Pattern 2: Multiplication Table for 5")
for i in range(1, 11):
    print(f"5 x {i} = {5 * i}")

# number between 1 and 50 is randomly selected
import random

# Set the secret number between 1 and 50
secret_number = random.randint(1, 50)

# Initialize the number of guesses
guess_count = 0

# While loop for the game
print("Welcome to the Number Guessing Game!")
print("I have selected a secret number between 1 and 50. Try to guess it!")

while True:
    guess = int(input("Enter your guess: "))
    guess_count += 1

    if guess < secret_number:
        print("Too low! Try again.")
    elif guess > secret_number:
        print("Too high! Try again.")
    else:
        print(f"Congratulations! You guessed the number in {guess_count} tries.")
        break

```

```python
# List of students with their details
students = [
    {
        "name": "Ahmed",
        "id": 1001,
        "grades": {
            "Math": 85,
            "Science": 90,
            "History": 88
        }
    },
    {
        "name": "Ali",
        "id": 1002,
        "grades": {
            "Math": 78,
            "Science": 82,
            "History": 75
        }
    },
    {
        "name": "Omer",
        "id": 1003,
        "grades": {
            "Math": 92,
            "Science": 95,
            "History": 91
        }
    }
]

# Calculate the average grade for a student
def calculate_avg(grades):
    return sum(grades.values()) / len(grades)

# Display all students with their grades
print("Student Grades:")
for student in students:
    print(f"{student['name']} (ID: {student['id']})")
    for course, grade in student['grades'].items():
        print(f"  {course}: {grade}")
    print()

# Display average grade for each student
print("Average Grade for Each Student:")
for student in students:
    avg = calculate_avg(student['grades'])
    print(f"{student['name']}: {avg:.2f}")
print()

# Find and display student with highest average grade
best_student = max(students, key=lambda x: calculate_avg(x['grades']))
print(f"Student with the highest average grade: {best_student['name']}")

# Display students with average grade above 80
print("\nStudents with an average grade above 80:")
for student in students:
    avg = calculate_avg(student['grades'])
    if avg > 80:
        print(f"{student['name']} with average grade {avg:.2f}")

# Calculate average grade for a specific course (Math in this case)
course_name = "Math"
course_grades = [student['grades'][course_name] for student in students if course_name in student['grades']]
average_course_grade = sum(course_grades) / len(course_grades)
print(f"\nAverage grade for {course_name}: {average_course_grade:.2f}")

# Add a new student
new_student = {
    "name": "Yusuf",
    "id": 1004,
    "grades": {
        "Math": 80,
        "Science": 85,
        "History": 87
    }
}

students.append(new_student)
print("\nNew student added:", new_student['name'])
```
